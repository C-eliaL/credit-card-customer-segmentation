```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] 
plt.rcParams['axes.unicode_minus'] = False
```

```python
# 載入原始數據
# 注意相對路徑的寫法 ../ 代表回到上一層資料夾
df = pd.read_csv('../data/BankChurners.csv') # 檔名雖然不符，但先照舊
print("信用卡客戶行為數據載入成功")
```

```python
# 數據清洗與前處理
# 處理缺失值，使用中位數填補
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].median(), inplace=True)
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].median(), inplace=True)
print("\n缺失值處理完成。")

# 刪除 CUST_ID，因為它對分群沒有幫助
df_cleaned = df.drop('CUST_ID', axis=1)
print("已移除 CUST_ID 欄位。")
```

```python
# 探索性數據分析 (EDA) 
# 設定圖表儲存路徑
FIGURES_PATH = "../reports/figures/"

# 繪製消費總額的分佈圖
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['PURCHASES'], bins=50, kde=True)
plt.title('客戶消費總額分佈')
plt.savefig(FIGURES_PATH + 'purchases_distribution.png')
plt.show()

# 繪製帳戶餘額的分佈圖
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['BALANCE'], bins=50, kde=True)
plt.title('客戶帳戶餘額分佈')
plt.savefig(FIGURES_PATH + 'balance_distribution.png')
plt.show()
```

```python
# 將乾淨的數據儲存起來，供後續的 Python 腳本使用
df_cleaned.to_csv('../data/processed/creditcard_cleaned.csv', index=False)
print("清洗後的數據已儲存至 data/processed/creditcard_cleaned.csv")
```
